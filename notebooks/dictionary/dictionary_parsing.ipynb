# Dictionary Parser — Usage Example

```python
from dictionary_parsing import parse_file,expand
from config import DICTIONARY_PATH

entries = parse_file(DICTIONARY_PATH)
print(f"Total entries: {len(entries)}")
expanded_entries = expand(entries,split_commas=True)
print(f"Expanded entries: {len(expanded_entries)}")
```

### Preview first 3 entries

```python
for e in expanded_entries[:15000]:
    label = e["transliteration"] 
    sense = f"[{e['sense']}]" if e["sense"] else ""
    translations = " | ".join(e["all_translations"])
    print(f"{label:<25} {sense:<5} {translations}")
```

### Save as JSONL for training

```python
import json

with open("dictionary_entries.jsonl", "w", encoding="utf-8") as f:
    for entry in entries:
        f.write(json.dumps(entry, ensure_ascii=False) + "\n")

print("Done.")
```
