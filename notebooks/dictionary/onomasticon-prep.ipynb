```python
import pandas as pd 
from config import DICTIONARY_DATASET_INPUTS
from utils import stack_csvs_from_folder
```

```python
df = pd.read_csv(DICTIONARY_DATASET_INPUTS/"onomasticon.csv")
```

```python
import pandas as pd

rows = []

for _, row in df.iterrows():
    # Parse spellings (semicolon-separated)
    spellings_raw = str(row['Spellings_semicolon_separated']) if pd.notna(row['Spellings_semicolon_separated']) else ''
    spellings = [s.strip() for s in spellings_raw.split(';') if s.strip()]

    # Parse aliases (semicolon-separated too, based on sample)
    aliases_raw = str(row['Aliases']) if pd.notna(row['Aliases']) else ''
    aliases = [a.strip() for a in aliases_raw.split(';') if a.strip()]

    name = str(row['Name']).strip()

    # One entry per (Name, Spelling)
    for spelling in spellings:
        rows.append({'english_name': name, 'spelling': spelling})

    # One entry per (Alias, Spelling)
    for alias in aliases:
        for spelling in spellings:
            rows.append({'english_name': alias, 'spelling': spelling})

result = pd.DataFrame(rows, columns=['english_name', 'spelling'])
result = result.drop_duplicates()
result.insert(0, 'id', range(1, len(result) + 1))

print("\nSample output:")
print(result.head(20).to_string())
```

```python
result.to_csv("onomasticon-cleaned.csv")
```
