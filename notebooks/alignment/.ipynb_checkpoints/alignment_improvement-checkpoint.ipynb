```python

def compute_length_stats(df: pd.DataFrame) -> pd.DataFrame:
    """Add word-count columns for transliteration and translation."""
    df = df.copy()
    df[Columns.TRANSLITERATION + "len"] = df[Columns.TRANSLITERATION].apply(
        lambda x: len(x)
    )
    df[Columns.TRANSLATION + "len"] = df[Columns.TRANSLATION].apply(lambda x: len(x))

    df["len_ratio"] = (
        df[Columns.TRANSLATION + "len"] / df[Columns.TRANSLITERATION + "len"]
    )
    return df

def print_length_summary(df: pd.DataFrame):
    print("\n--- Length Summary ---")
    print(f"  Total rows:              {len(df)}")
    print(
        f"  Avg transliteration len: {df[Columns.TRANSLITERATION + 'len'].mean():.1f} words"
    )
    print(
        f"  Avg translation len:     {df[Columns.TRANSLATION + 'len'].mean():.1f} words"
    )
    print(f"  Avg len ratio (t/t):     {df['len_ratio'].mean():.2f}")
    print(f"  Std len ratio:           {df['len_ratio'].std():.2f}")


def find_suspicious_rows(df: pd.DataFrame, z_threshold: float = 2.0) -> pd.DataFrame:
    """
    Flag rows where the length ratio is more than z_threshold standard deviations
    from the mean. These are likely misaligned or garbage pairs.
    """
    mean = df["len_ratio"].mean()
    std = df["len_ratio"].std()

    lower = mean - z_threshold * std
    upper = mean + z_threshold * std

    suspicious = df[(df["len_ratio"] < lower) | (df["len_ratio"] > upper)].copy()
    suspicious["z_score"] = (suspicious["len_ratio"] - mean) / std

    print(f"\n--- Suspicious Rows (z > {z_threshold}) ---")
    print(f"  Mean ratio:   {mean:.2f}")
    print(f"  Lower bound:  {lower:.2f}")
    print(f"  Upper bound:  {upper:.2f}")
    print(
        f"  Flagged rows: {len(suspicious)} / {len(df)}  ({len(suspicious) / len(df):.1%})"
    )

    return suspicious.sort_values("z_score", key=abs, ascending=False)


def print_examples(suspicious: pd.DataFrame, n: int = 10):
    """Print the most extreme misaligned pairs for manual inspection."""
    print(f"\n--- Top {n} Most Misaligned Pairs ---")
    for i,(_, row) in enumerate(suspicious.head(n).iterrows()):
        print(f"{i}  Ratio: {row['len_ratio']:.2f}  (z={row['z_score']:.2f}). ID: {str(row["id"])}")
        print(
            f"{i}  Translit [{row[Columns.TRANSLITERATION + 'len']} words]: {str(row[Columns.TRANSLITERATION])[:120]}"
        )
        print(
            f"{i}  Transl   [{row[Columns.TRANSLATION + 'len']} words]: {str(row[Columns.TRANSLATION])[:120]}"
        )
        print()


def length_distribution(df: pd.DataFrame, buckets=(0.2, 0.5, 1.0, 2.0, 3.0, 5.0)):
    """Show what fraction of pairs fall in each ratio bucket."""
    print("\n--- Ratio Distribution ---")
    prev = 0.0
    for b in buckets:
        count = ((df["len_ratio"] > prev) & (df["len_ratio"] <= b)).sum()
        print(f"  {prev:.1f} < ratio <= {b:.1f} : {count:>6}  ({count / len(df):.1%})")
        prev = b
    count = (df["len_ratio"] > prev).sum()
    print(f"  ratio > {prev:.1f}           : {count:>6}  ({count / len(df):.1%})")

def plot_length_analysis(df: pd.DataFrame, z_threshold: float = 2.0):
    """
    Visualizes the length relationships and identifies outliers.
    """
    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    translit_col = Columns.TRANSLITERATION + "len"
    transl_col = Columns.TRANSLATION + "len"
    mean_ratio = df["len_ratio"].mean()
    std_ratio = df["len_ratio"].std()

    # 1. Scatter Plot: Transliteration vs Translation Length
    sns.scatterplot(
        data=df, 
        x=translit_col, 
        y=transl_col, 
        alpha=0.5, 
        ax=axes[0],
        hue="len_ratio", 
        palette="viridis"
    )
    
    # Add a line representing the mean ratio
    x_vals = range(int(df[translit_col].max()))
    axes[0].plot(x_vals, [mean_ratio * x for x in x_vals], color='red', linestyle='--', label=f'Mean Ratio ({mean_ratio:.2f})')
    
    axes[0].set_title("Length Correlation")
    axes[0].set_xlabel("Transliteration Length (Words/Chars)")
    axes[0].set_ylabel("Translation Length (Words/Chars)")
    axes[0].legend()

    # 2. Histogram: Distribution of Length Ratios
    sns.histplot(df["len_ratio"], bins=50, kde=True, ax=axes[1], color="teal")
    
    # Highlight the Z-score thresholds
    lower = mean_ratio - z_threshold * std_ratio
    upper = mean_ratio + z_threshold * std_ratio
    axes[1].axvline(lower, color='orange', linestyle='--', label=f'-{z_threshold}σ ({lower:.2f})')
    axes[1].axvline(upper, color='orange', linestyle='--', label=f'+{z_threshold}σ ({upper:.2f})')
    
    axes[1].set_title("Distribution of Length Ratios")
    axes[1].set_xlabel("Length Ratio (Translation / Transliteration)")
    axes[1].legend()

    plt.tight_layout()
    plt.savefig("length_analysis_plot.png")
    print("\nPlot saved as 'length_analysis_plot.png'")
```

```python
tp = TextProcessor()
pre_success_df = success_df.copy()
pre_success_df['transliteration'] = success_df['transliteration'].apply(
    lambda x: tp.preprocess_transliteration_text(
        x, 
        named_determinatives=False, 
        with_hyphens=False, 
        normalize_chars=False, 
        separate_compounds=False
    )
)
```

```python
success_df = compute_length_stats(success_df)
# Summary stats
print_length_summary(success_df)
# Distribution across ratio buckets
length_distribution(success_df)
```

```python
pre_success_df = compute_length_stats(pre_success_df)
# Summary stats
print_length_summary(pre_success_df)
# Distribution across ratio buckets
length_distribution(pre_success_df)
```

```python
z_threshold = 0.2

# Find suspicious rows
suspicious = find_suspicious_rows(pre_success_df, z_threshold=z_threshold)

# Print the worst offenders for manual inspection
print_examples(suspicious, n=30)
```

```python
plot_length_analysis(success_df,z_threshold=z_threshold)
```
