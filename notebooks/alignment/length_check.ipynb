# Maas aligned data

```python
import pandas as pd
from transformers import AutoTokenizer
from data_processing.processing import TextProcessor

df = pd.read_csv("../../dataset/maas_aligned.csv")
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")

translations = df["translation"].dropna().tolist()
transliterations = df["transliteration"].dropna().tolist()

trans_lengths = [len(tokenizer(t).input_ids) for t in translations]
translit_lengths = [len(tokenizer(t).input_ids) for t in transliterations]

print("=== TRANSLATION (target) ===")
print(f"Max:    {max(trans_lengths)}")
print(f"Mean:   {sum(trans_lengths)/len(trans_lengths):.1f}")
print(f"95th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.95)]}")
print(f"99th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.99)]}")

print("\n=== TRANSLITERATION (source) ===")
print(f"Max:    {max(translit_lengths)}")
print(f"Mean:   {sum(translit_lengths)/len(translit_lengths):.1f}")
print(f"95th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.95)]}")
print(f"99th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.99)]}")



print("AFTER PROCESSING")




tp = TextProcessor()

df["transliteration"] = df["transliteration"].apply(
            lambda x: tp.preprocess_transliteration_text(
                        transliteration_text=x,
                        separate_compounds=False,
                        with_hyphens=False, 
                        named_determinatives=True,
                        normalize_chars=True)
            )

df["translation"] = df["translation"].apply(
                lambda x: tp.preprocess_translation_text(x))


# Drop nulls just in case
translations = df["translation"].dropna().tolist()
transliterations = df["transliteration"].dropna().tolist()


# Get lengths
trans_lengths = [len(tokenizer(t).input_ids) for t in translations]
translit_lengths = [len(tokenizer(t).input_ids) for t in transliterations]

import matplotlib.pyplot as plt

print("=== TRANSLATION (target) ===")
print(f"Max:    {max(trans_lengths)}")
print(f"Mean:   {sum(trans_lengths)/len(trans_lengths):.1f}")
print(f"95th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.95)]}")
print(f"99th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.99)]}")

print("\n=== TRANSLITERATION (source) ===")
print(f"Max:    {max(translit_lengths)}")
print(f"Mean:   {sum(translit_lengths)/len(translit_lengths):.1f}")
print(f"95th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.95)]}")
print(f"99th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.99)]}")

# Plot distributions
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(trans_lengths, bins=50)
axes[0].set_title("Translation lengths")
axes[0].set_xlabel("Tokens")
axes[1].hist(translit_lengths, bins=50)
axes[1].set_title("Transliteration lengths")
axes[1].set_xlabel("Tokens")
plt.tight_layout()
plt.show()
```

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from transformers import AutoTokenizer
from src.data_processing.processing import TextProcessor
from src.data_processing.alignment import Aligner

df = pd.read_csv("../dataset/training_input/external/akkademia.csv")
tokenizer = AutoTokenizer.from_pretrained("google/byt5-base")

translations = df["translation"].dropna().tolist()
transliterations = df["transliteration"].dropna().tolist()

trans_lengths = [len(tokenizer(t).input_ids) for t in translations]
translit_lengths = [len(tokenizer(t).input_ids) for t in transliterations]

print("=== TRANSLATION (target) ===")
print(f"Max:    {max(trans_lengths)}")
print(f"Mean:   {sum(trans_lengths)/len(trans_lengths):.1f}")
print(f"95th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.95)]}")
print(f"99th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.99)]}")

print("\n=== TRANSLITERATION (source) ===")
print(f"Max:    {max(translit_lengths)}")
print(f"Mean:   {sum(translit_lengths)/len(translit_lengths):.1f}")
print(f"95th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.95)]}")
print(f"99th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.99)]}")



print("AFTER PROCESSING")




tp = TextProcessor()

df["transliteration"] = df["transliteration"].apply(
            lambda x: tp.preprocess_transliteration_text(
                        transliteration_text=x,
                        separate_compounds=False,
                        with_hyphens=False, 
                        named_determinatives=True,
                        normalize_chars=True)
            )

df["translation"] = df["translation"].apply(
                lambda x: tp.preprocess_translation_text(x))


# Drop nulls just in case
translations = df["translation"].dropna().tolist()
transliterations = df["transliteration"].dropna().tolist()


# Get lengths
trans_lengths = [len(tokenizer(t).input_ids) for t in translations]
translit_lengths = [len(tokenizer(t).input_ids) for t in transliterations]

import matplotlib.pyplot as plt

print("=== TRANSLATION (target) ===")
print(f"Max:    {max(trans_lengths)}")
print(f"Mean:   {sum(trans_lengths)/len(trans_lengths):.1f}")
print(f"95th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.95)]}")
print(f"99th %: {sorted(trans_lengths)[int(len(trans_lengths)*0.99)]}")

print("\n=== TRANSLITERATION (source) ===")
print(f"Max:    {max(translit_lengths)}")
print(f"Mean:   {sum(translit_lengths)/len(translit_lengths):.1f}")
print(f"95th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.95)]}")
print(f"99th %: {sorted(translit_lengths)[int(len(translit_lengths)*0.99)]}")

# Plot distributions
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(trans_lengths, bins=50)
axes[0].set_title("Translation lengths")
axes[0].set_xlabel("Tokens")
axes[1].hist(translit_lengths, bins=50)
axes[1].set_title("Transliteration lengths")
axes[1].set_xlabel("Tokens")
plt.tight_layout()
plt.show()
```
